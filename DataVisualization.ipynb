```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "C:\\Users\\Capaciti\\Downloads\\student_scores.csv"
df = pd.read_csv(file_path)

# Calculate the average scores for each subject
avg_scores = df[["Math", "Science", "English", "History"]].mean()

# Creating the pie chart
plt.figure(figsize=(6, 6))
plt.pie(avg_scores, labels=avg_scores.index, autopct="%1.1f%%", colors=["blue", "green", "red", "orange"])
plt.title("Average Scores by Subject")
plt.show()

# Creating the bar chart
plt.figure(figsize=(8, 5))
df.set_index("Name")[["Math", "Science", "English", "History"]].plot(kind="bar", figsize=(10, 6))
plt.title("Student Scores by Subject")
plt.ylabel("Scores")
plt.xlabel("Students")
plt.xticks(rotation=45)
plt.legend(title="Subjects")
plt.show()
```
